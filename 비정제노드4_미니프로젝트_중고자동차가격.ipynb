## ì›í•«ì¸ì½”ë”©

```python
# ë²”ì£¼í˜• ì»¬ëŸ¼ë“¤ì— ì›-í•« ì¸ì½”ë”©ì„ ì ìš©í•©ë‹ˆë‹¤.

car_df = pd.get_dummies(car_df, drop_first = True)
```

```python
from sklearn.preprocessing import RobustScaler
```

```python
# Q. Robust Scalerë¥¼ rs ë³€ìˆ˜ë¡œ ì •ì˜í•´ì¤ë‹ˆë‹¤.

rs = RobustScaler()
```

```python
# Q. robust scalingì„ ì ìš©í•œ ë°ì´í„°í”„ë ˆìž„ì„ car_df ë³€ìˆ˜ì— ë‹¤ì‹œ ì €ìž¥í•©ë‹ˆë‹¤.
car_rs_df = rs.fit_transform(car_df)
```

```python
from sklearn.decomposition import PCA
```

```python
# Q. ì£¼ì„±ë¶„ì´ 5ê°œì¸ PCA ëª¨ë¸ì„ ì •ì˜í•©ë‹ˆë‹¤.
pca = PCA(5)
```

```python
# Q. car_df ë°ì´í„°ì— PCAë¥¼ ì ìš©í•©ë‹ˆë‹¤.

pd.DataFrame(pca.fit_transform(car_rs_df))
```

```python
# Q. ì›ë³¸ ë°ì´í„° ëŒ€ë¹„ PCA ë°ì´í„°ì˜ ë¶„ì‚°(explained variance) ë¹„ìœ¨ì„ ê³„ì‚°í•©ë‹ˆë‹¤.

pca.explained_variance_ratio_
```

```python
# ì£¼ì„±ë¶„ì˜ ê°œìˆ˜ê°€ 2~10ê°œì¼ ë•Œ explained variance ratioë¥¼ ê°ê° ê³„ì‚°í•©ë‹ˆë‹¤.

for i in range(2, 11):
    pca = PCA(i)
    pca.fit(car_rs_df)
    print(i, round(pca.explained_variance_ratio_.sum(), 2))
```

```python
pca = PCA(7)
```
